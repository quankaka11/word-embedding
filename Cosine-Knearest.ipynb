```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import gensim
```

# Cosine

```python
import re

def clean_whitespace_in_file(input_path, output_path):
    with open(input_path, 'r', encoding='utf-8') as infile, \
         open(output_path, 'w', encoding='utf-8') as outfile:

        for i, line in enumerate(infile, start=1):
            # Dùng regex để loại bỏ mọi khoảng trắng thừa (bao gồm tab và non-breaking spaces)
            cleaned_line = re.sub(r'\s+', ' ', line.strip())

            # Kiểm tra và ghi lại dòng nếu không rỗng
            if cleaned_line:
                outfile.write(cleaned_line + '\n')
            else:
                print(f"Bỏ qua dòng {i}: Dòng rỗng hoặc chỉ chứa khoảng trắng.")

input_file = 'W2V_150.txt'
output_file = 'W2V_150_cleaned.txt'

clean_whitespace_in_file(input_file, output_file)
```

```python
from gensim.models import KeyedVectors
import os

path = 'W2V_150_cleaned.txt'
# if os.path.isfile(path):
#     print("File ton tai")
# else:
#     print("----")
model = KeyedVectors.load_word2vec_format(path, binary=False)
```

```python
def get_word_vector(model, word):
    try:
        return model[word] #trả về biểu diễn
    except KeyError:
        return np.zeros(model.vector_size) # trả về vector 0 nếu không tìm thấy từ
```

```python
from numpy.linalg import norm
def cosine_similarity_vectors(word1, word2):
    vec1 = get_word_vector(model,word1)
    vec2 = get_word_vector(model,word2)
    return np.dot(vec1,vec2)/(norm(vec1)*norm(vec2))
```

```python
df = pd.read_csv('datasets/Visim-400/Visim-400.txt', sep='\t')
```

```python
for i in range(0,400):
    word1 = df['Word1'][i]
    word2 = df['Word2'][i]
    similarity = cosine_similarity_vectors(word1,word2)
    print(f"Cosine similarity of {word1} and {word2} is: {similarity}")
```

# K - nearest words

```python
def find_k_nearest_neighbors(word, k):
    words = model.index_to_key
    word_vec = get_word_vector(model, word)
    word_vec = word_vec.reshape(1, -1)
    words_vec = np.array([get_word_vector(model, w) for w in words])
    cos_sim = cosine_similarity(word_vec, words_vec)
    cos_sim = cos_sim.reshape(-1)
    cos_sim_idx = np.argsort(cos_sim)
    k_nearest = [words[i] for i in cos_sim_idx[-k-1:-1]]
    return k_nearest[::-1]

word = 'code'
k = 5
k_nearest = find_k_nearest_neighbors(word, k)
print(f"{k} từ gần với từ '{word}': {k_nearest}")
```
